```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Gerar dados contínuos
np.random.seed(0)  # Para reprodutibilidade
grupo1 = np.random.normal(loc=50, scale=10, size=100)  # Média 50, desvio padrão 10
grupo2 = np.random.normal(loc=55, scale=10, size=100)  # Média 55, desvio padrão 10
```

```python
# Realizar o teste t de Student
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
```

```python
# Resultados do teste
print(f'Estatística t: {t_stat}')
print(f'Valor p: {p_value}')
```

Interpretação dos Resultados

    A estatística t indica a diferença entre as médias em relação à variabilidade dos dados.
    
    O valor p ajuda a determinar a significância estatística da diferença observada. Um valor p menor que 0.05 geralmente indica que a diferença entre os grupos é estatisticamente significativa.


```python
# Visualizar os dados
plt.figure(figsize=(10, 6))
plt.hist(grupo1, alpha=0.5, label='Grupo 1', bins=20)
plt.hist(grupo2, alpha=0.5, label='Grupo 2', bins=20)
plt.axvline(np.mean(grupo1), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(grupo2), color='orange', linestyle='dashed', linewidth=1)
plt.title('Distribuição dos Grupos')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()
plt.show()

# um histograma para visualizar a distribuição dos dados
# de ambos os grupos e estão marcadas as médias com linhas tracejadas.
```
