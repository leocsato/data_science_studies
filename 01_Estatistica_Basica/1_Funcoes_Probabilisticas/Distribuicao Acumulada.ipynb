```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Gerar dados de uma distribuição normal
np.random.seed(0)  # Para reprodutibilidade
dados = np.random.normal(loc=0, scale=1, size=1000)  # Média 0, desvio padrão 1
```

```python
# Calcular a CDF empírica
dados_sorted = np.sort(dados)
n = len(dados)
cdf_empirica = np.arange(1, n + 1) / n
```

```python
# Calcular a CDF teórica da distribuição normal
x = np.linspace(-4, 4, 1000)
cdf_teorica = stats.norm.cdf(x, loc=0, scale=1)
```

```python
# Plotar a CDF empírica e a CDF teórica
plt.figure(figsize=(10, 6))
plt.plot(dados_sorted, cdf_empirica, label='CDF Empírica', color='blue')
plt.plot(x, cdf_teorica, label='CDF Teórica (Normal)', color='orange', linestyle='dashed')
plt.title('Função de Distribuição Acumulada (CDF)')
plt.xlabel('Valor')
plt.ylabel('Probabilidade Acumulada')
plt.legend()
plt.grid()
plt.show()
```

```python
# Realizar o teste de Kolmogorov-Smirnov
ks_stat, p_value = stats.kstest(dados, 'norm', args=(0, 1))
```

```python
# Resultados do teste
print(f'Estatística KS: {ks_stat}')
print(f'Valor p: {p_value}')
```

Interpretação dos Resultados

    O gráfico permite visualizar como a CDF empírica se compara à CDF teórica.
    
    A estatística KS indica a maior diferença entre as duas distribuições.
    
    O valor p ajuda a determinar a significância do teste. Um valor p menor que 0.05 geralmente indica que os dados não seguem a distribuição normal.
