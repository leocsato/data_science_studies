```python
import numpy as np
import scipy
from scipy.stats import norm
import seaborn as sns
```

```python
norm.cdf(np.array([-1., 0, 1]))
```

```python
# Variáveis aleatórias de uma distribuição beta
variaveis = scipy.stats.beta.rvs(5, 5, loc=0, scale=1, size=500)
```

```python
# (a, b, loc, scale)
params = scipy.stats.beta.fit(variaveis)
```

```python
params
```

```python
# Calcula CDF
x = np.linspace(0, 1, 500)
cdf = norm.cdf(x, params[0:2])
```

```python
sns.lineplot(x=variaveis, y=cdf)
```
