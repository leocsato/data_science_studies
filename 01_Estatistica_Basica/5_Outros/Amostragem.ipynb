```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Definindo a semente para reprodutibilidade
np.random.seed(42)
```

```python
# Gerando uma população normal (por exemplo, alturas em cm)
populacao = np.random.normal(loc=170, scale=10, size=1000)
# média 170, desvio padrão 10, 1000 indivíduos
```

```python
# Função para realizar amostragem
def realizar_amostragem(populacao, tamanho_amostra):
    return np.random.choice(populacao, size=tamanho_amostra, replace=False)
```

```python
# Parâmetros da amostragem
tamanho_amostra = 30
amostra = realizar_amostragem(populacao, tamanho_amostra)
```

```python
# Cálculo de estatísticas básicas
media_amostra = np.mean(amostra)
desvio_padrao_amostra = np.std(amostra, ddof=1)  # ddof=1 para amostra
n = len(amostra)
```

```python
# Exibindo resultados
print(f"Média da amostra: {media_amostra:.2f}")
print(f"Desvio padrão da amostra: {desvio_padrao_amostra:.2f}")
```

```python
# Teste t para a média da amostra
media_populacao = np.mean(populacao)
t_stat, p_value = stats.ttest_1samp(amostra, media_populacao)

print(f"Estatística t: {t_stat:.2f}, Valor p: {p_value:.4f}")
```

```python
# Visualizando a população e a amostra
plt.figure(figsize=(12, 6))
plt.hist(populacao, bins=30, alpha=0.5, label='População', color='blue')
plt.hist(amostra, bins=30, alpha=0.7, label='Amostra', color='orange')
plt.axvline(media_amostra, color='orange', linestyle='dashed', linewidth=2, label='Média da Amostra')
plt.axvline(media_populacao, color='blue', linestyle='dashed', linewidth=2, label='Média da População')
plt.title('Distribuição da População e Amostra')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```
