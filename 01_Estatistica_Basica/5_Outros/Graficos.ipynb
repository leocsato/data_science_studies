```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definindo a semente para reprodutibilidade
np.random.seed(42)

# Gerando um conjunto de dados fictício
n = 100
dados = {
    'Altura': np.random.normal(loc=170, scale=10, size=n),  # Altura em cm
    'Peso': np.random.normal(loc=70, scale=15, size=n),     # Peso em kg
    'Idade': np.random.randint(18, 65, size=n)               # Idade em anos
}

# Criando um DataFrame
df = pd.DataFrame(dados)

# Estatísticas descritivas
print(df.describe())
```

# Gráfico de dispersão (scatter plot) entre Altura e Peso

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Altura'], df['Peso'], alpha=0.6)
plt.title('Gráfico de Dispersão: Altura vs Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid()
plt.show()
```

# Histograma da Altura

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Altura'], bins=20, kde=True, color='blue', alpha=0.6)
plt.title('Histograma da Altura')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
plt.grid()
plt.show()
```

# Boxplot do Peso por faixa etária

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=pd.cut(df['Idade'], bins=[18, 25, 35, 45, 55, 65]), y='Peso', data=df)
plt.title('Boxplot do Peso por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Peso (kg)')
plt.grid()
plt.show()
```

# Gráfico de correlação

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Matriz de Correlação')
plt.show()
```
