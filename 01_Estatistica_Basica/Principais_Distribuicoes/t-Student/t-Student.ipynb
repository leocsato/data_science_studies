A distribuição t-Student é comumente usada em testes de hipóteses quando o tamanho da amostra é pequeno e o desvio padrão da população é desconhecido.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
def plot_t_distribution(x,
                        degrees_of_freedom,
                        mean=0,
                        std_dev=1,
                        range_multiple=3):
    """
    Traçar o gráfico da distribuição t-Student com determinados graus de liberdade.
    
    Parâmetros:
    degrees_of_freedom (int): Graus de liberdade para a distribuição t-Student.
    mean (float): Média da distribuição (o padrão é 0).
    std_dev (float): Desvio padrão da distribuição (o padrão é 1).
    range_multiple (float): Número de desvios padrão a serem incluídos no intervalo do gráfico (o padrão é 3).
    """
    y = t.pdf(x, df=degrees_of_freedom, loc=mean, scale=std_dev)

    plt.plot(x, y, label=f't-Distribution (df={degrees_of_freedom})')
    plt.title(f't-Student Distribution (df={degrees_of_freedom})')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
x = np.linspace(-5, 5, 100)
```

```python
degrees_of_freedom = 2
plot_t_distribution(x, degrees_of_freedom)
```

```python
degrees_of_freedom = 5
plot_t_distribution(x, degrees_of_freedom)
```

```python
degrees_of_freedom = 10
plot_t_distribution(x, degrees_of_freedom)
```

```python
degrees_of_freedom = 25
plot_t_distribution(x, degrees_of_freedom)
```
