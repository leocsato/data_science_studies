## Teste de uma distribuição para verificar se ela é normalmente distribuída

```python
import pandas as pd

np.random.seed(0)
df = pd.DataFrame(np.random.normal(0, 1.5, (100, 1)),
                  columns=['data'])
```

```python
from scipy.stats import normaltest
k2, p = normaltest(df.data)
```

```python
p
```

A função testa a hipótese nula de que os dados são provenientes de uma distribuição normal. 
Quanto maior for o valor de p, maior será a probabilidade de que os dados sejam provenientes de uma distribuição normal. 
Aplica um limite (geralmente 0,05) acima do qual a hipótese nula não pode ser rejeitada.

## Gráfico

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

range = np.arange(-3,3,0.001)
plt.plot(range, norm.pdf(range, 0, 1))
plt.show()
```

```python
def draw_z_score(x, cond, mu, sigma, title):
    y = norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x, y)
    plt.fill_between(z, 0, norm.pdf(z, mu, sigma))
    plt.title(title)
    plt.show()
```

```python
x = np.arange(-3,3,0.001)
z0 = -0.75
draw_z_score(x, x<z0, 0, 1, 'z <-0.75')
```

```python
x = np.arange(-3,3,0.001)
z0 = 0.75
draw_z_score(x, x > z0, 0, 1, ' z > 0.75')
```

```python
x = np.arange(-3,3,0.001)
z0 = 0.75
draw_z_score(x, (-z0 < x) & (x < z0), 0, 1, '-0.75 < z <0.75')
```

# Exemplo 1

Exemplo prático da distribuição normal:

Suponha que você está estudando as alturas de uma amostra de 100 estudantes de uma escola.

Após analisar os dados, você descobriu que a média das alturas é de 170 cm e o desvio padrão é de 5 cm.

Você quer saber qual a probabilidade de um estudante selecionado aleatoriamente ter uma altura entre 165 cm e 175 cm.

Para resolver esse problema, podemos utilizar a distribuição normal.

Qual é a probabilidade (em %)?

```python
import numpy as np
from scipy.stats import norm
```

```python
media = 170
```

```python
desvio = 5
```

```python
norm.cdf(175, loc = media, scale = desvio) - norm.cdf(165, loc = media, scale = desvio)
```
