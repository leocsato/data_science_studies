A distribuição geométrica fornece a chance de sucesso na tentativa n.

O sucesso na primeira tentativa é p.

O sucesso em n é (1-p)^(n-1)*p. 

Em Python, a operação ^ é realizada por **.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import geom
```

```python
prob = 0.3
N = 15
p = np.zeros(N)
d = np.zeros(N)
for k in range(1,N+1):
    p[k-1] = (1-prob)**(k-1)*prob
    d[k-1] = geom.pmf(k,prob)
plt.bar(range(1,N+1),p, color = 'gray')
plt.xlim(1,N+1)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('Primeiro sucesso em N')
plt.show()

print('Probabilidade de sucesso depois de', N ,'tentativas',
      (1-sum(p)),'\nde acordo com a fórmula.\n')
print('Probabilidade de sucesso depois de', N ,'tentativas',
      (1-sum(d)),'\nde acordo com a geom.pmf')
```

# Exemplo 1

Exemplo prático da Distribuição Geométrica:
Suponha que você está jogando um jogo de cara ou coroa com uma moeda justa. Você quer saber a probabilidade de obter a primeira cara após 3 lançamentos.

Neste caso, temos:

p = 0,5 (probabilidade de obter cara em cada lançamento, pois a moeda é justa)

X = 3 (número de tentativas necessárias para obter a primeira cara)

Qual é a Probabilidade (em %)?

```python
prob = 0.5
N = 3
p = np.zeros(N)
d = np.zeros(N)
for k in range(1,N+1):
    p[k-1] = (1-prob)**(k-1)*prob
    d[k-1] = geom.pmf(k,prob)
plt.bar(range(1,N+1),p, color = 'gray')
plt.xlim(1,N+1)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('Primeiro sucesso em N')
plt.show()

print('Probabilidade de sucesso depois de', N ,'tentativas',
      (1-sum(p)),'\nde acordo com a fórmula.\n')
print('Probabilidade de sucesso depois de', N ,'tentativas',
      (1-sum(d)),'\nde acordo com a geom.pmf')
```
