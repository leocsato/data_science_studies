```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
# Definindo a probabilidade de sucesso
p = 0.7  # Probabilidade de sucesso
```

```python
# Gerando amostras da distribuição de Bernoulli
n_samples = 1000
samples = bernoulli.rvs(p, size=n_samples)
```

```python
# Calculando a frequência de sucessos e fracassos
success_count = np.sum(samples)
failure_count = n_samples - success_count
```

```python
# Resultados
print(f'Número de sucessos (1): {success_count}')
print(f'Número de fracassos (0): {failure_count}')
```

```python
# Calculando a probabilidade de sucesso e fracasso
prob_success = bernoulli.pmf(1, p)
prob_failure = bernoulli.pmf(0, p)
```

```python
print(f'Probabilidade de sucesso (1): {prob_success:.4f}')
print(f'Probabilidade de fracasso (0): {prob_failure:.4f}')
```

```python
# Visualizando os resultados
plt.figure(figsize=(8, 5))
plt.bar(['Sucesso (1)', 'Fracasso (0)'], [success_count, failure_count], color=['green', 'red'])
plt.title('Resultados da Distribuição de Bernoulli')
plt.ylabel('Frequência')
plt.xlabel('Resultado')
plt.show()
```
