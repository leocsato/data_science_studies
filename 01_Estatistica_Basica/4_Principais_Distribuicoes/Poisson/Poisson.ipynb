-----

Uma variável aleatória X que tem uma distribuição de Poisson representa o 
número de eventos que ocorrem em um intervalo de tempo fixo com uma taxa 
λ.

λ informa a taxa na qual o número de eventos ocorre.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
n = np.arange(0, 10)
rate = 2
print(n)
```

```python
y = stats.poisson.pmf(n, rate)
print(y)
```

```python
plt.plot(n, y, 'o-')
plt.title('Poisson: λ =%i' % rate)
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidade do Número de Eventos')
plt.show()
```

Observa-se que o número de eventos atinge o pico em torno da média. 

No geral, você pode esperar um número λ de eventos. 

Tente valores diferentes de λ e n e veja como a forma da distribuição muda.

Próximo teste:

1.000 variáveis aleatórias de uma distribuição de Poisson 

```python
data = stats.poisson.rvs(mu = 2, loc = 0, size = 500)
print("Média: %g" %np.mean(data))
print("Desvio-Padrão: %g" % np.std(data, ddof=1))

plt.figure()
plt.hist(data, bins=9)
plt.xlim(0, 10)
plt.xlabel("número de eventos")
plt.title("Simulando variáveis aleatórias de Poisson")
plt.show()
```
