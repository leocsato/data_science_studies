A distribuição binomial é uma distribuição de probabilidade discreta que descreve o número de sucessos em uma sequência de experimentos de Bernoulli independentes, cada um com a mesma probabilidade de sucesso. É frequentemente usada em situações onde você está contando o número de sucessos em um número fixo de tentativas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import scipy.stats as stats
```

# Exemplo

Se a moeda for jogada seis vezes, a probabilidade de obter três caras é máxima,

enquanto que obter uma única cara ou cinco caras tem a menor probabilidade.

Isso pode ser calculado usando o código abaixo em python.

```python
fig, ax = plt.subplots(1, 1)
x = range(7)
n, p = 6, 0.5
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,label='Probabilidade de Successo')          
ax.legend(loc='best', frameon=False)          
plt.show()
```

-----

Suponha que um dado seja jogado 5 vezes.

Qual é a probabilidade de obter exatamente 2 quatros?

A probabilidade de qualquer resultado é 1/6=0,167. 

Nosso código teria a seguinte aparência.

```python
fig, ax = plt.subplots(1, 1)
x = range(6)
n, p = 5, 0.166
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,label='Probabilidade')          
ax.legend(loc='best', frameon=False)          
plt.show()
```

# Outro exemplo

Exemplo de distribuição binomial: Qual é a probabilidade de obter 2 caras em 10 jogadas de uma moeda justa? 

Nesse experimento, a probabilidade de obter uma cara é 0,3,  
Isso significa que, em média, você pode esperar que 3 lançamentos da moeda sejam cara. 

Defino todos os valores possíveis que o lançamento da moeda pode assumir: k = np.arange(0,11)
Pode-se observar: zero cara, uma cara e até dez caras. 
Usado stats.binom.pmf para calcular a função de massa de probabilidade para cada observação. 
Retorna uma lista de 11 elementos, esses elementos representam a probabilidade associada a cada observação.

```python
n = 10
p = 0.3
k = np.arange(0, 21)
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k, binomial, 'o-')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade de Sucessos', fontsize = 15)
plt.show()
```

```python
binom_sim = data = stats.binom.rvs(n = 10, p = 0.3, size = 10000)
print( "Média: %g" % np.mean(binom_sim))
print( "Desvio Padrão: %g" % np.std(binom_sim, ddof = 1) )
plt.hist(binom_sim, bins = 10,) # normed
plt.xlabel("x")
plt.ylabel("Densidade")
plt.show()
```

# Outro Exemplo #3

```python
# Definindo os parâmetros da distribuição binomial
n = 10  # Número de tentativas
p = 0.5  # Probabilidade de sucesso em cada tentativa
```

```python
# Gerando amostras da distribuição binomial
n_samples = 1000
samples = binom.rvs(n, p, size=n_samples)
```

```python
# Calculando a frequência de sucessos
success_counts = np.arange(0, n + 1)
frequencies = [np.sum(samples == k) for k in success_counts]
```

```python
# Resultados
print(f'Frequências de sucessos em {n} tentativas:')
for k, freq in zip(success_counts, frequencies):
    print(f'Número de sucessos {k}: {freq}')
```

```python
# Calculando a probabilidade de um número específico de sucessos
k = 5  # Número de sucessos desejado
prob_k_successes = binom.pmf(k, n, p)
print(f'Probabilidade de obter exatamente {k} sucessos em {n} tentativas: {prob_k_successes:.4f}')
```

```python
# Visualizando os resultados
plt.figure(figsize=(10, 6))
plt.bar(success_counts, frequencies, color='blue', alpha=0.7, label='Frequência')
plt.title('Distribuição Binomial: Frequência de Sucessos')
plt.xlabel('Número de Sucessos')
plt.ylabel('Frequência')
plt.xticks(success_counts)
plt.legend()
plt.grid(axis='y')
plt.show()
```
