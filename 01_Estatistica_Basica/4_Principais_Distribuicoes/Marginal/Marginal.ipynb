A distribuição marginal refere-se à distribuição de uma ou mais variáveis em um conjunto de dados, ignorando as outras variáveis. Em outras palavras, é a distribuição de uma variável individual obtida a partir de uma distribuição conjunta.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exemplo

Para ilustrar a distribuição marginal, vamos usar um exemplo simples com duas variáveis. Vamos gerar dados para duas variáveis aleatórias e, em seguida, calcular e visualizar a distribuição marginal de uma delas.

```python
# Gerando dados de exemplo para duas variáveis
np.random.seed(0)  # Para reprodutibilidade
x = np.random.normal(loc=5, scale=1, size=1000)  # Variável X
y = np.random.normal(loc=10, scale=2, size=1000)  # Variável Y
```

```python
# Criando um DataFrame para facilitar a visualização
data = np.vstack((x, y)).T
```

```python
# Visualizando a distribuição conjunta
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.kdeplot(x=x, y=y, cmap='Blues', fill=True, thresh=0, levels=20)
plt.title('Distribuição Conjunta de X e Y')
plt.xlabel('X')
plt.ylabel('Y')

# Calculando e visualizando a distribuição marginal de X
plt.subplot(1, 2, 2)
sns.histplot(x, bins=30, kde=True, color='orange')
plt.title('Distribuição Marginal de X')
plt.xlabel('X')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Observação:

Neste exemplo, a distribuição marginal de X é obtida simplesmente ignorando Y. A distribuição marginal é útil para entender o comportamento de uma variável individual em um conjunto de dados multivariados. Você pode aplicar o mesmo conceito para calcular a distribuição marginal de Y ou de qualquer outra variável em um conjunto de dados.
