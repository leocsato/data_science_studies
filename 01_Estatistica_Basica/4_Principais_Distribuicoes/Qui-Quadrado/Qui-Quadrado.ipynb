A distribuição qui-quadrado é uma distribuição de probabilidade que é frequentemente usada em testes de hipóteses, especialmente em testes de independência e em análises de variância. A distribuição qui-quadrado é definida por um único parâmetro, que é o número de graus de liberdade.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
# Definindo os parâmetros da distribuição qui-quadrado
df = 5  # Graus de liberdade
```

```python
# Gerando dados de exemplo
n_samples = 1000
samples = chi2.rvs(df, size=n_samples)
```

```python
# Calculando a função de densidade de probabilidade (PDF)
x = np.linspace(0, 20, 1000)  # Valores de x para calcular a PDF
pdf = chi2.pdf(x, df)
```

```python
# Visualizando os resultados
plt.figure(figsize=(10, 6))

# Histograma das amostras
plt.hist(samples,
         bins=30,
         density=True,
         alpha=0.6,
         color='g',
         label='Histograma das Amostras')

# Gráfico da PDF
plt.plot(x,
         pdf,
         'r-',
         lw=2,
         label='PDF da Distribuição Qui-Quadrado')

# Configurações do gráfico
plt.title('Distribuição Qui-Quadrado')
plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid()
plt.show()
```

Obs:

O número de graus de liberdade é definido como 5, mas você pode modificar esse valor para ver como isso afeta a distribuição.
