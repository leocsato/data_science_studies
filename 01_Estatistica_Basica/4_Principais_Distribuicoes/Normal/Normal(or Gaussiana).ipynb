A distribuição normal, também conhecida como distribuição gaussiana, é uma das distribuições de probabilidade mais importantes em estatística. Ela é caracterizada por sua forma de sino e é definida por dois parâmetros: a média (μ) e o desvio padrão (σ).

A distribuição normal é frequentemente usada para modelar fenômenos naturais e sociais.

```python
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Teste de uma distribuição para verificar se ela é normalmente distribuída

```python
import pandas as pd

np.random.seed(0)
df = pd.DataFrame(np.random.normal(0, 1.5, (100, 1)),
                  columns=['data'])
```

```python
from scipy.stats import normaltest
k2, p = normaltest(df.data)
```

A função testa a hipótese nula de que os dados são provenientes de uma distribuição normal. 
Quanto maior for o valor de p, maior será a probabilidade de que os dados sejam provenientes de uma distribuição normal. 
Aplica um limite (geralmente 0,05) acima do qual a hipótese nula não pode ser rejeitada.

# Exemplo #1

```python
# Definindo os parâmetros da distribuição normal
mu = 0      # Média
sigma = 1   # Desvio padrão
```

```python
# Gerando dados de exemplo
n_samples = 1000
samples = np.random.normal(mu, sigma, n_samples)
```

```python
# Calculando a função de densidade de probabilidade (PDF)
x = np.linspace(-4, 4, 1000)  # Valores de x para calcular a PDF
pdf = norm.pdf(x, mu, sigma)
```

```python
# Visualizando os resultados
plt.figure(figsize=(10, 6))

# Histograma das amostras
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Histograma das Amostras')

# Gráfico da PDF
plt.plot(x, pdf, 'r-', lw=2, label='PDF da Distribuição Normal')

# Configurações do gráfico
plt.title('Distribuição Normal')
plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid()
plt.show()
```

# Exemplo #2

Exemplo prático da distribuição normal:

Suponha que você está estudando as alturas de uma amostra de 100 estudantes de uma escola.

Após analisar os dados, você descobriu que a média das alturas é de 170 cm e o desvio padrão é de 5 cm.

Você quer saber qual a probabilidade de um estudante selecionado aleatoriamente ter uma altura entre 165 cm e 175 cm.

Para resolver esse problema, podemos utilizar a distribuição normal.

Qual é a probabilidade (em %)?

```python
# Definindo os parâmetros da distribuição normal
mu = 170  # Média das alturas
sigma = 5  # Desvio padrão das alturas
```

```python
# Definindo os limites de altura
lower_limit = 165
upper_limit = 175
```

```python
# Calculando a probabilidade de estar entre os limites
probability = norm.cdf(upper_limit, mu, sigma) - norm.cdf(lower_limit, mu, sigma)
```

```python
# Convertendo a probabilidade para porcentagem
probability_percentage = probability * 100
```

```python
# Resultados
msg = f"""A probabilidade de um estudante ter uma altura entre {lower_limit} cm
          e {upper_limit} cm é de {probability_percentage:.2f}%."""
print(msg)
```
