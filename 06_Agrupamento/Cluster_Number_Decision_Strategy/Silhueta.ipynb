A Silhueta é um método de interpretação e validação da consistência em clusters de dados.

Ela fornece uma representação gráfica de quão bem cada objeto foi classificado e mede a semelhança de um objeto com seu próprio cluster em comparação com outros clusters.

---

O valor da silhueta varia de -1 a +1, em que um valor alto indica que o objeto tem boa correspondência com seu próprio cluster e pouca correspondência com os clusters vizinhos.

A Silhueta pode ser calculada com qualquer métrica de distância, como a distância euclidiana ou a distância de Manhattan.

O Método da Silhueta é usado para encontrar o número ideal de clusters em um conjunto de dados e é considerado melhor do que o método do cotovelo.

Desvantagens:
Com o aumento da dimensionalidade dos dados, fica difícil atingir valores tão altos devido à maldição da dimensionalidade, pois as distâncias se tornam mais semelhantes.

A Medida da Silhueta é especializada para medir a qualidade do cluster quando os clusters têm formato convexo e pode não ter um bom desempenho se os clusters de dados tiverem formatos irregulares ou tamanhos variados.

----

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
np.random.seed(0)
```

```python
x1 = np.random.randint(0, 10 ,21)
x2 = np.random.randint(0, 10 ,21)
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
```

```python
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Base de Dados')
plt.scatter(x1, x2)
plt.show()
```

```python
# Lista para armazenar os valores do Silhouette Score
silhouette_scores = []
```

```python
# Tentar diferentes números de clusters
for n_clusters in range(2, 18):
    # Treinar o modelo de clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(X)
    
    # Calcular o Silhouette Score
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
    print(f"Para n_clusters={n_clusters}, a Silhueta é {silhouette_avg}")
```

```python
# Plotar o gráfico do Silhouette Score em função do número de clusters
plt.plot(range(2, 18),
         silhouette_scores,
         marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Medida da Silhueta')
plt.title('Medida da Silhueta x Número de clusters')
plt.show()
```
