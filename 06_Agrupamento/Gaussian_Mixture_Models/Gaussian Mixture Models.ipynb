Os Modelos de Mistura Gaussiana (GMMs) são um tipo de algoritmo de agrupamento usado para agrupar dados não rotulados com base em sua similaridade. Eles envolvem a combinação de várias distribuições normais, cada uma identificada por k ∈ {1,..., K}, em que K é o número de clusters no conjunto de dados.

Diferentemente do K-means, os Modelos de Mistura Gaussiana (GMMs) levam em conta a variação e retornam a probabilidade de um ponto de dados pertencer a cada um dos K clusters.

Os GMMs podem lidar até mesmo com clusters muito oblongos e realizar uma classificação suave, o que significa que eles fornecem probabilidades de que um determinado ponto de dados pertença a cada um dos possíveis clusters.

O algoritmo envolve a estimativa dos valores das variáveis latentes γ, a atualização dos valores dos parâmetros (ou seja, média e coeficientes de mistura) e a repetição do processo até a convergência do modelo.

----

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import StratifiedKFold
```

```python
X, _ = make_blobs(n_samples=500,
                  centers=7,
                  cluster_std=0.5,
                  random_state=0)
```

```python
X
```

```python
gmm = GaussianMixture(n_components=7, random_state=0)
gmm.fit(X)
```

```python
labels = gmm.predict(X)
```

```python
plt.scatter(X[:, 0],
            X[:, 1],
            c=labels,
            s=40,
            cmap='viridis')
plt.title('GMM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
