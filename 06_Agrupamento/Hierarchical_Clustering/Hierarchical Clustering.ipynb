É um método de análise e construção de clusters que possui duas estratégias: top-down e bottom-up. Top-down é o método divisivo, enquanto Bottom-up é o método chamado aglomerativo.
O nome do método é Agrupamento Hierárquico.

O método aglomerativo começa com cada observação em seu próprio cluster, depois vai combinando pares de clusters similares de acordo com uma métrica de distância entre os clusters.

No método divisivo todos começam em um grande cluster, que vai sendo quebrado em clusters menores conforme eles descem na hierarquia.

```python
import time
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
model = AgglomerativeClustering(n_clusters=5)
model.fit(X)
labels = model.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1],
            c=labels, cmap='viridis', s=50, alpha=0.7)
plt.title('Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Single

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograma (Hierarquia dos Clusters)")
dendrogram(linkage(X, method='single'),
       truncate_mode='level', p=3)
plt.xlabel("Amostras")
plt.ylabel("Distância")
plt.show()
```

# Full

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograma (Hierarquia dos Clusters)")
dendrogram(linkage(X, method='complete'),
       truncate_mode='level', p=3)
plt.xlabel("Amostras")
plt.ylabel("Distância")
plt.show()
```

# Average

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograma (Hierarquia dos Clusters)")
dendrogram(linkage(X, method='average'),
       truncate_mode='level', p=3)
plt.xlabel("Amostras")
plt.ylabel("Distância")
plt.show()
```

# Centroid

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograma (Hierarquia dos Clusters)")
dendrogram(linkage(X, method='centroid'),
       truncate_mode='level', p=3)
plt.xlabel("Amostras")
plt.ylabel("Distância")
plt.show()
```

# Ward

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograma (Hierarquia dos Clusters)")
dendrogram(linkage(X, method='ward'),
       truncate_mode='level', p=3)
plt.xlabel("Amostras")
plt.ylabel("Distância")
plt.show()
```
