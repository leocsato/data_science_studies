Em termos de gráficos, a diferença entre a regularização de Lasso e Ridge pode ser ilustrada através das curvas de contorno das funções de perda associadas a esses métodos de regularização.

    Regularização de Lasso (L1):
        Adiciona o termo de penalização L1, que é a soma dos valores absolutos dos coeficientes.
        As curvas de contorno da função de perda para Lasso parecem polígonos regulares com vértices nos eixos. Isso significa que as soluções com diferentes valores dos coeficientes que têm a mesma penalização estão em vértices desses polígonos, onde um ou mais coeficientes são zero.
        
    Regularização de Ridge (L2):
        A função de perda adiciona o termo de penalização L2, que é a soma dos quadrados dos coeficientes.
        As curvas de contorno da função de perda parecem círculos centrados na origem. Isso significa que as soluções com diferentes valores dos coeficientes que têm a mesma penalização estão a uma distância fixa do ponto (0,0) no espaço de coeficientes.

Em resumo:
A regularização L2 tende a espalhar a penalização igualmente entre todos os coeficientes, enquanto a regularização L1 tem a tendência de gerar soluções esparsas, com alguns coeficientes zerados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo a grade de valores para os coeficientes
x = np.linspace(-2, 2, 300)
y = np.linspace(-2, 2, 300)
X, Y = np.meshgrid(x, y)
```

```python
# Função de perda para as Regularizações
ridge_loss = X**2 + Y**2
lasso_loss = np.abs(X) + np.abs(Y)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 2)
plt.contour(X, Y, lasso_loss, levels=np.linspace(0, 4, 10))
plt.title("Lasso (L1)")
```

```python
plt.subplot(1, 2, 1)
plt.contour(X, Y, ridge_loss, levels=np.linspace(0, 4, 10))
plt.title("Ridge (L2)")
```

```python
plt.tight_layout()
plt.show()
```
